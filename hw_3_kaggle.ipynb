```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import ColumnTransformer
```

```python
data['TotalSpent'] = data['TotalSpent'].replace(' ', '')
data['TotalSpent'] = pd.to_numeric(data['TotalSpent'])
data['TotalSpent'] = data['TotalSpent'].fillna(data['TotalSpent'].median())
```

```python
X = data[feature_cols]
y = data[target_col]

# 
preprocessor = ColumnTransformer([
    ("num", StandardScaler(), num_cols),
    ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols)
])

# Pipeline  LogisticRegressionCV
pipe = make_pipeline(
    preprocessor,
    LogisticRegressionCV(
        Cs=[100, 10, 1, 0.1, 0.01, 0.001],
        scoring='roc_auc',
        cv=5,
        n_jobs=-1,
        max_iter=1000,
        refit=True
    )
)

# 
pipe.fit(X, y)

#  C
best_C = pipe.named_steps['logisticregressioncv'].C_[0]
print(" C:", best_C)

# ROC-AUC   
roc_auc = pipe.score(X, y)
print(f"ROC-AUC: {roc_auc:.4f}")
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

X = data[feature_cols]
y = data[target_col]

X_train, X_test, y_train,  y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state = 42)


model = CatBoostClassifier(
    iterations=2000,
    learning_rate=0.1,
    depth=5,
    eval_metric='AUC',
    verbose=100
)

model.fit(X_train, y_train, cat_features=cat_cols, eval_set=(X_test, y_test), early_stopping_rounds=50)

print(f'ROC-AUC {roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])}')
```
